```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
conversion_rate = 83.95
```

```python
df = pd.read_csv("D:\Projects\Practice\postgreSQL\Project_SQL\my_CSV_Files\\1_top_paying_jobs.csv")
```

```python
df
```

```python
df['salary_inr'] = df['salary_year_avg'] * conversion_rate
```

```python
sort_data = df.sort_values(by='salary_inr', ascending=False)
```

```python
companies_sort = sort_data['company_name']
rupee_salaries_sorted = sort_data['salary_inr']
```

```python
fig, ax = plt.subplots(figsize=(10, 10), facecolor='grey')  # Set figure background color
ax.set_facecolor('grey')
ax.barh(companies_sort.head(12), rupee_salaries_sorted.head(12), color='darkgrey')
ax.set_xlabel('Salaries (in Crore)', fontweight = 'bold', fontsize = '14')
ax.set_ylabel('Companies', fontweight = 'bold', fontsize = '14')
ax.set_title('Average Salary Distribution for Top 10 Paying Data Scientist Jobs In 2023 In India', fontweight = 'bold', fontsize = '14')
plt.gca().invert_yaxis()
plt.show()
```

```python
df_2 = pd.read_csv('D:\Projects\Practice\postgreSQL\Project_SQL\my_CSV_Files\\2_top_paying_job_skills.csv')
```

```python
df_2
```

```python
conversion_rate = 83.95
```

```python
df_2['salary_inr'] = df_2['salary_year_avg'] * conversion_rate
```

```python
sort_data_2 = df_2.sort_values(by='salary_inr', ascending=False)
```

```python
skills_2 = sort_data_2['skills']
rupee_salaries_sorted = sort_data_2['salary_inr']
```

```python
fig, ax = plt.subplots(figsize=(10, 10), facecolor='grey')  # Set figure background color
ax.set_facecolor('grey')
bars = ax.barh(skills_2.head(10), rupee_salaries_sorted.head(10), color='darkgrey')
ax.set_xlabel('Salaries (in Crore)', fontweight = 'bold', fontsize = '14')
ax.set_ylabel('Skills', fontweight = 'bold', fontsize = '14')
ax.set_title('Skill Count for Top 10 Paying Data Scientist Jobs In India (2023) ', fontweight = 'bold', fontsize = '14')
plt.gca().invert_yaxis()
ax.bar_label(bars, fmt='%.2f', label_type='edge', color='black', fontsize=12)
plt.show()
```
